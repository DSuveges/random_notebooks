````python
import json
import pandas as pd
import gzip

oldFile = '/Users/dsuveges/project_data/ot/evidence_input/21.04/uniprot/cttv011-05-02-2021.json.gz'
d= []
with gzip.open(oldFile, 'r') as f:
    for line in f:
        data = json.loads(line)
        d.append({
            'target': data['target']['id'].split('/')[-1],
            'disease': data['disease']['id'].split('/')[-1],
            'datasourceId': data['sourceID']
        })

old_df = pd.DataFrame(d)
````

````python
print(f'Number of evidence: {len(old_df)}')

print(f'Number of disease: {len(old_df.disease.unique())}')
print(f'Number of targets: {len(old_df.target.unique())}')
print(f'Number of target/disese pairs: {len(old_df[["disease", "target"]].drop_duplicates())}')
````

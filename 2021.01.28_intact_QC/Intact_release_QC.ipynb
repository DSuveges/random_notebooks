# QC Intact data release


OTAR044 data release linked to every Itact release. But once the graphQL dump is generated, we do a little bit of tests to make sure everything looks good.

```python
%%bash
for release in "2020-11-25" "2021-01-18"; do
    echo "Fetching release: ${release}"
    URL="ftp://ftp.ebi.ac.uk/pub/databases/intact/various/ot_graphdb/${release}/data/interactor_pair_interactions.json"
    wget $URL -O interactor_pair_interactions_${release}.json
    wc -l interactor_pair_interactions_${release}.json
done

```

```python
import pandas as pd
import json

# We read two datasets at a time so we can compare the numbers: odd trends can be pointed out:
oldRelease = '2020-11-25'
newRelease = '2021-01-18'

# Read first file:
jsonDocuments = []
with open(f'interactor_pair_interactions_{newRelease}.json', 'r') as f:
    for line in f:
        jsonDocuments.append(json.loads(line))

    
df = pd.json_normalize(jsonDocuments)
```

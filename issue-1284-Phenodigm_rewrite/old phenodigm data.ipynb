````python
import pyspark.sql
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark import SparkContext
import re

global spark

# SparkContext.setSystemProperty('spark.executor.memory', '20g')

spark = (pyspark.sql.SparkSession
    .builder
    .appName("phenodigm_parser")
     .config("spark.driver.bindAddress", "localhost")
    .getOrCreate()
)

phenodigmFile = '/Users/dsuveges/project_data/ot/evidence_input/20.11/phenodigm/phenodigm-29-10-2020.json'

ph_df = (
    spark.read.json(phenodigmFile)
    .select(col('target.id').alias('target'),col('disease.id').alias('disease'),
            'unique_association_fields.disease_phenodigm_id','unique_association_fields.model_description',
           col('evidence.biological_model.model_id'), col())
)

# Print answers:
print(ph_df.count())
print(ph_df.select('target').distinct().count())
print(ph_df.select('disease').distinct().count())
````
